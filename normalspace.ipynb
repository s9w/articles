# Generating numbers with spacing density of normal distribution

We want to generate N points on a range between x_min and x_max, with both ends included. In other words: $[x_0, x_1, \dots, x_{N-1}]$ with $x_0=x_\text{min}$ and $x_{N-1}=x_\text{max}$. This is trivial to do with a uniform distribution, and NumPy even has a log distribution builtin.

But sometimes you want to generate more points around an interesting position. Ideally, their density would be proportional to a normal distribution. There are often builtin functions to generate random numbers in such a way, but what about a controlled number range? For examples how this will look in the end, see the end of this notebook.

We'll assume a normal definition of the normal distribution with a standard deviation of $\sigma$ and a center of $\mu$:

$$\frac{1}{\sigma \sqrt{2\pi}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

Let's have a look:

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['axes.grid'] = True
mpl.rcParams['legend.labelspacing'] = 0.0
```

```python
def normal(x, mu, sigma):
    return 1.0/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2.0/(2.0*sigma**2.0))

x = np.linspace(-3, 5, 100)
parameters = [(1.0, 1.0), (1.0, 0.5), (0.0, 0.7)]
for mu, sigma in parameters:
    plt.plot(x, normal(x, mu, sigma), label="$\\mu$={} $\\sigma$={}".format(mu, sigma));
plt.legend(loc="upper right");
```

The key idea for our normal range is that the points divide a normal distribution into segments of equal areas. So first we need the area.

Normal distributions are always normalized, so the integral from $-\infty$ to $\infty$ is 1. But we want to look at the general case of a range from some x_min to x_max, not the whole range. That's given by the cumulative distribution function (CDF) $\Phi$, which gives us the integral from $-\infty$ to some point. For the normal distrubution, that is

$$\Phi = \frac 1 2 \left(1+\text{erf}\left(\frac{x-\mu}{\sigma\sqrt 2}\right)\right)$$

That expressions contains the error function, which can be found in NumPy, Boost for C++ or even natively since C++11.

```python
from scipy.special import erf, erfinv
def phi(x, mu, sigma):
    return 0.5*(1.0+erf((x-mu)/(sigma*np.sqrt(2.0))))

x = np.linspace(-2, 3, 100)
mu=0.0; sigma=0.7
plt.plot(x, normal(x, mu, sigma), label="Normal dist")
plt.plot(x, phi(x, mu, sigma), label="CDF")
plt.legend(loc="upper right");
```

Now the more general case where we have a range:

```python
ax = plt.subplot(111)
x_min = -0.3; x_max = 1.3
x = np.linspace(-1, 2, 100)
plt.plot(x, normal(x, mu, sigma))
plt.fill_between(x, 0, normal(x, mu, sigma), where=(x_max>x)&(x>x_min), alpha=0.2);
```

The area under the normal dist between our x_min and x_max is given by $A=\Phi_\text{max}-\Phi_\text{min}$. We'll divide that into N-1 parts with `A0=A/(N-1)` and generate our sequence from there. But for that we need to solve the equation for $x$:

$$x=\sigma\sqrt 2\cdot\text{erf}^{-1}(2\Phi-1)+\mu$$

The CDF of our clipped range is given by `phi_clip = phi_min + i*A0` where `i` iterates from 0 to N-1. The cases `i=0` and `i=N-1` are identical to x_min and x_max.

```python
def normalSpace(x_min, x_max, N, mu, sigma):
    phi_min = phi(x_min, mu, sigma)    
    A = phi(x_max, mu, sigma) - phi_min
    A0 = A/(N-1)
    seq = [np.sqrt(2)*sigma*erfinv((phi_min+i*A0)*2-1)+mu for i in range(0,N)]
    return np.asarray(seq)

N = 6
x_normal = normalSpace(x_min, x_max, N, mu, sigma)
plt.plot(x, normal(x, mu, sigma), "-", color="gray")
plt.plot(x_normal, normal(x_normal, mu, sigma), "o");
```

This works like a charm. More examples:

```python
def plot_normals(x_min, x_max, N, mu, sigma):
    X = normalSpace(x_min, x_max, N, mu, sigma)
    plt.plot(X, normal(X, mu, sigma), marker="o", markersize=5, ls="", color="black", alpha=0.7)
    X = np.linspace(x_min, x_max)
    plt.plot(X, normal(X, mu, sigma), color="black", alpha=0.5)

plt.figure(figsize=(7,3))
ax = plt.subplot(121)
plot_normals(-3, 1, N=11, mu=-1.5, sigma=0.8)
plot_normals(-3, 1, N=11, mu=0.4, sigma=1.3)

ax = plt.subplot(122)
plot_normals(0,3, N=15, mu=0.0, sigma=0.9)
plot_normals(0,3, N=10, mu=0.0, sigma=1.5)
```
